```python
filename = 'how_long_in_step.pkl'
with open(filename, 'rb') as f: 
    time_on_steps = pickle.load(f) 
time_on_steps
```

```python
#rename columns to be descriptive

time_on_steps.rename(columns = {0:'visit_id',1:'client_id',2:'process_step',3:'date_time',4:'next_step_time', 5:'time_spent_sec', 6:'Variation'},inplace=True)
time_on_steps
```

```python
# Step 1: Filter DataFrame for 'start' steps
start_step = time_on_steps[time_on_steps['process_step'] == '0']

# Step 2: Group by 'client_id'
grouped = start_step.groupby('client_id')

# Step 3: Calculate the average time spent for each client
average_time_spent_start = grouped['time_spent_sec'].mean()

average_time_start = pd.DataFrame(average_time_spent_start).reset_index()
average_time_start
```
