```python
filename = 'how_long_in_step.pkl'
with open(filename, 'rb') as f: 
    time_on_steps = pickle.load(f) 
time_on_steps
```

```python
#rename columns to be descriptive

time_on_steps.rename(columns = {0:'visit_id',1:'client_id',2:'process_step',3:'date_time',4:'next_step_time', 5:'time_spent_sec', 6:'Variation'},inplace=True)
time_on_steps
```

```python
# Step 1: Filter DataFrame for 'start' steps
start_step = time_on_steps[time_on_steps['process_step'] == '0']

# Step 2: Group by 'client_id'
grouped = start_step.groupby('client_id')

# Step 3: Calculate the average time spent for each client
average_time_spent_start = grouped['time_spent_sec'].mean()

average_time_start = pd.DataFrame(average_time_spent_start).reset_index()
average_time_start.describe()
```

```python
# Step 1: Filter DataFrame for 'step 1' steps
step_1 = time_on_steps[time_on_steps['process_step'] == '1']

# Step 2: Group by 'client_id'
grouped = step_1.groupby('client_id')

# Step 3: Calculate the average time spent for each client
average_time_spent_1 = grouped['time_spent_sec'].mean()

average_time_step_1 = pd.DataFrame(average_time_spent_1).reset_index()
average_time_step_1.describe()
```

```python
# Step 1: Filter DataFrame for 'step 2' steps
step_2 = time_on_steps[time_on_steps['process_step'] == '2']

# Step 2: Group by 'client_id'
grouped = step_2.groupby('client_id')

# Step 3: Calculate the average time spent for each client
average_time_spent_2 = grouped['time_spent_sec'].mean()

average_time_step_2 = pd.DataFrame(average_time_spent_2).reset_index()
average_time_step_2.describe()
```

```python
# Step 1: Filter DataFrame for 'step 3' steps
step_3 = time_on_steps[time_on_steps['process_step'] == '3']

# Step 2: Group by 'client_id'
grouped = step_3.groupby('client_id')

# Step 3: Calculate the average time spent for each client
average_time_spent_3 = grouped['time_spent_sec'].mean()

average_time_step_3 = pd.DataFrame(average_time_spent_3).reset_index()
average_time_step_3.describe()
```

```python
time_spent_on_confirm = time_on_steps[time_on_steps['process_step'] == '4'][['client_id','process_step','time_spent_sec']]
time_spent_on_confirm
```
